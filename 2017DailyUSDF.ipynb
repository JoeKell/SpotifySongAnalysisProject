```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Not every CSV requires an encoding, but be aware this can come up
US_Daily_2017_df = pd.read_csv("Resources/2017-daily-us-charts.csv")
```

```python
# Show just the header
US_Daily_2017_df.head()
```

```python
Deaths_df = pd.read_csv("Resources/2017 Musician Deaths (Wiki).csv")
Deaths_df
```

```python
merged_df = pd.merge(Deaths_df, US_Daily_2017_df, how="left", on=["Artist", "Artist"])
merged_df
                                                                
```

```python
drop_df = merged_df.dropna()
drop_df
```

```python
# drop_df.to_csv("Resources/Cleaned Death.csv",encoding="utf-8", index=False, header=True)
```

```python
drop_columns_df = drop_df.drop(columns=['Age ','Position','ID'])
drop_columns_df
```

```python
chuck_berry_df = drop_columns_df.loc[drop_columns_df["Name"] == "Chuck Berry",:]
chuck_berry_df.head()
```

```python
streams_sum = chuck_berry_df.groupby(["Artist", "Date"])
streams_sum.aggregate(np.sum).head()
```

```python
chuckberry = chuck_berry_df.groupby("Date")
chuckberry_streams = pd.DataFrame(chuckberry["Streams"].count())

chuckberry_chart = chuckberry_streams.plot(kind='line', title="Chuck Berry Streams After Death", color="red")
chuckberry_chart.set_xlabel("Date")
chuckberry_chart.set_ylabel("Streams")

plt.show()
```
